# Experimentación del TP

````python
#Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def nlogn(x):
    return x * np.log(x);
fig = plt.figure()
````

### Instancias random

Primero podemos ver la correlación de nuestras medidas con la cota teórica que estimamos. Para esto usamos 

```
sns.lmplot
```

que hace una regresión lineal sobre las dos columnas que le pasamos.

````python
df_random = pd.read_csv("runtime_tp_random.csv")
df_random['cota'] = nlogn(df_random['n'])

ax1 = sns.lmplot(x='time', y='cota', data=df_random)
plt.xlabel("tiempo (s)")
plt.ylabel("cota teorica")
````

Este gráfico tiene el problema de que los n que consideramos crecen exponencialmente. Podemos hacer el plot con escala logaritmica para que se vea mejor.

````python
df_random['log_time'] = np.log(df_random['time'])
df_random['log_cota'] = np.log(df_random['cota'])
ax1 = sns.lmplot(x='log_time', y='log_cota', data=df_random)
plt.xlabel("log_tiempo (s)")
plt.ylabel("log_cota")
````

Incluso podemos calcular el coeficiente de Pearson, que es una medida de dependencia lineal entre variables. Mientras más cerca esté de 1, más confianza tenemos de que nuestras variables están linealmente correlacionadas.

````python
r = np.corrcoef(df_random['log_time'], df_random['log_cota'])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
````

### Comparación entre instancias

Vamos a hacer plots mostrando la diferencia de tiempo entre los tres casos que propusimos.

````python
df_random['time_sorted'] = df_sorted['time']
df_random['time_unsorted'] = df_unsorted['time']

fig, ax = plt.subplots()

ax.scatter(df_random['n'],df_random['time'],color='yellow', label='random')
ax.scatter(df_random['n'],df_random['time_sorted'],color='blue', label='sorted')
ax.scatter(df_random['n'],df_random['time_unsorted'],color='red', label='unsorted')
plt.xlabel("n")
plt.ylabel("tiempo (s)")

ax.legend()
ax.grid(True)

plt.show()
````
